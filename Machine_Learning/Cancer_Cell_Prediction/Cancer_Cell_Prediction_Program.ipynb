IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import seaborn as sb
```

READING DATA

```python
df = pd.read_csv('Cancer_Cell_Prediction_Data.csv')
df.head()
```

CHECKING FOR NULL VALUES

```python
df.isnull().sum()
```

VISUALIZING THE COUNTS OF CANCER CELL TYPES USING BARPLOT

```python
cell_types = df['diagnosis'].groupby(df['diagnosis']).count()
sb.barplot(x = cell_types.index, y = cell_types.values)
```

CHANGING THE VALUES OF THE DIAGNOSIS COLUMN TO NUMERIC

```python
df['diagnosis'] = df['diagnosis'].map({'B' : 0, 'M' : 1})
df.head()
```

REMOVING THE UNWANTED ID COLUMN USING DEL

```python
del df['id']
df.head()
```

SPLITTING THE DATA INTO X & Y

```python
x = df.drop(columns = ['diagnosis'], axis = 1)
y = df['diagnosis']

x.head()
```

```python
y.head()
```

SPLITTING THE X & Y INTO TRAINING & TESTING DATA SETS

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

CREATING A DECISION_TREE_CLASSIFIER MODEL & TRAINING IT ON THE TRAINING DATA SET

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()

model.fit(x_train, y_train)
```

PREDICTING ON THE TESTING DATA SET

```python
y_predict = model.predict(x_test)
y_predict
```

SEEING HOW THE MODEL PERFORMED

```python
from sklearn.metrics import confusion_matrix, accuracy_score

print('CM :', confusion_matrix(y_test, y_predict))
print('AS :', accuracy_score(y_test, y_predict))
```

ADDING CROSS_VALIDATION TO REDUCE OVERFITTING

```python
from sklearn.model_selection import cross_val_score

for i in range(1, 21):
  model = DecisionTreeClassifier(max_depth = i)
  model.fit(x_train, y_train)

  y_predict = model.predict(x_train)
  ac = accuracy_score(y_train, y_predict)

  cs = cross_val_score(model, X = x_train, y = y_train, cv = 10)

  print('D : %2d    AC : %6.2f %%    CS : %13.10f' % (i, ac * 100, np.mean(cs)))
```
